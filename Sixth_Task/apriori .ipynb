## Data Preprocessing

```python
dataset = pd.read_csv('outputfile.csv', header = None)
transactions = []
for i in range(0, 7348):
  transactions.append([str(dataset.values[i,j]) for j in range(0, 11)])
```

```python
dataset[0].nunique()
```

## Training the Apriori model on the dataset

```python
from apyori import apriori#All these hyperparameters will depend on the business case study
rules = apriori(transactions = transactions, min_support = 0.003, min_confidence = 0.2, min_lift = 3, min_length = 2, max_length = 2)
```

## Visualising the results

### Displaying the first results coming directly from the output of the apriori function

```python
results = list(rules)
```

```python
results
```

### Putting the results well organised into a Pandas DataFrame

```python
def inspect(results):
    lhs         = [tuple(result[2][0][0])[0] for result in results]
    rhs         = [tuple(result[2][0][1])[0] for result in results]
    supports    = [result[1] for result in results]
    confidences = [result[2][0][2] for result in results]
    lifts       = [result[2][0][3] for result in results]
    return list(zip(lhs, rhs, supports, confidences, lifts))
resultsinDataFrame = pd.DataFrame(inspect(results), columns = ['Left Hand Side', 'Right Hand Side', 'Support', 'Confidence', 'Lift'])
```

### Displaying the results non sorted

```python
resultsinDataFrame.nlargest(n = 10, columns = 'Lift')
```
